# b. Outlier Treatment using the Z-score method

```python
import matplotlib.pyplot as mtp
for i in df:
    if df[i].dtype!='object':
        mtp.boxplot(df[i])
        mtp.show()
```

```python

for i in df:
    lower=(-3*df[i].std())+df[i].mean()
    upper=(3*df[i].std())+df[i].mean()
    df['zscore']=(df[i]-df[i].mean())/df[i].std()
    df[i]=np.where(df['zscore']>3,upper,df[i])
    df[i]=np.where(df['zscore']<-3,lower,df[i])
    df.drop(['zscore'],axis=1,inplace=True)
```

# QUESTION 3

## Perform dimensionality reduction using PCA such that the 95% of the variance is explained

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_sc=scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
```

```python
df_pca=pca.fit_transform(df_sc)
print('Original dataset shape')
print(df_sc.shape)
print('PCA dataset shape')
print(df_pca.shape)
```

# QUESTION 4

## Find the optimum value of k for k-means clustering using the elbow method. Plot the elbow curve

```python
from sklearn.cluster import KMeans
inert=[]
k=range(5,10)
for i in range(5,10):
    kmeans=KMeans(n_clusters=i)
    kmeans=kmeans.fit(df_pca)
    inert.append(kmeans.inertia_)
    
```

```python
mtp.figure(figsize=(15,10))
mtp.plot(k,inert,'bx-')
```

## from the elbow method, I found that the number of clusters =7 will give the optimum value as it gives minimum inertia

# QUESTION 5

## Find the optimum value of k for k-means clustering using the silhouette score method and specify the number of observations in each cluster using a bar plot

```python
from sklearn.metrics import silhouette_samples, silhouette_score
k=range(5,10)
for i in range(5,10):
    kmeans=KMeans(n_clusters=i)
    kmeans=kmeans.fit(df_pca)
    score = silhouette_score(df_pca, kmeans.labels_, metric='euclidean')
    print('Silhouetter Score of %d : %.3f' % (i,score))
    print('barchart to denote the number of observations in each cluster')
    l=range(0,i)
    mtp.hist(kmeans.labels_)
    mtp.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
fig, ax = mtp.subplots(3, 3, figsize=(15,8))
k=range(5,10)
for i in range(5,10):
    km = KMeans(n_clusters=i)
    q, mod = divmod(i, 3)
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(df_pca)
```

# from silhouette score, I have found that number of cluster 7 is found to be the optimal
